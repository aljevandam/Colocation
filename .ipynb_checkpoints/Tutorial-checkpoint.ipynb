### Import the data, and create a contingency table

````python
df = pd.read_csv('tutorial_data.csv') #import the example data set (consisting of BLS major occupation groups) 
q = df.pivot(index = 'city', columns = 'occupation', values = '0') #make contingency table q 
````

### Then define the class, choosing prior and prior weight

````python
Q = CoLoc(q, prior = 'uniform', nr_prior_obs = q.size) #define CoLoc class with a uniform prior, choosing the nr_prior_obs such that 1 pseudocount is added to each cell 
````

The location probabilities are their variance are available as properties of the class, as well as the shape of data: 
```python
Q.pci
Q.varpci 
Q.pi
Q.pc
Q.q_tot
Q.Ni 
Q.Nc
```

The matrix of pseudocounts is also available, and indeed is given by all ones since we used a uniform prior: 

````python
Q.alpha
````

The matrix of counts including the pseudocounts (i.e. q + alpha) is given by 
```python 
Q.qt 
```

Now, we can measures based on the location associations the standard deviations of their posterior. The following are available:
```python
Q.make_PMIpci() 
Q.make_stdPMIpci()
Q.make_KLpc_i()
Q.make_stdKLpc_i()
Q.make_KLpi_c() 
Q.make_stdKLpi_c() 
Q.make_MIpci() 
Q.make_stdMIpci()
```
For example:

````python
Q.make_PMIpci() ##compute location associations 
````

````python
Q.make_KLpc_i() ## compute localizations
````

## Co-location

Co-location measures are computed in the same way:
```python
Q.make_PMIpij() 
Q.make_stdPMIpij()
Q.make_KLpj_i()
Q.make_stdKLpj_i() 
Q.make_MIpij() 
```



For co-location, the properties 
```python
Q.pij
Q.varpij
Q.pj_i
Q.varpj_i
```

are available only after running 
```python
Q.make_colocprobs() 
``` 
or computing a measure based on these quantities (which runs make_colocprobs() automatically)

### Significance of associations 

The significance test is implemented for PMIpci and PMIpij. Return either significant or insignificant associations using:
```python
Q.make_sigPMIpci()
Q.make_insigPMIpci() 

Q.make_sigPMIpij()
Q.make_insigPMIij()
```

For example, to give only the significant co-location associations: 

````python
Q.make_sigPMIpij() ##non-significant associations are returned as NaN 
````
