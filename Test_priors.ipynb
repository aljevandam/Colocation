# Load data 

```python
#major occupation groups 
df = pd.read_csv('./BLS2016_clean.csv')

df = df.loc[df['OCC_GROUP']=='major']

df['OCC_TITLE'] = df['OCC_TITLE'].apply(lambda x: x[:-12])

#drop Farming and drop PR 
df = df[~df['OCC_CODE'].str.startswith('45')]
df = df[~df['AREA_NAME'].str.contains('PR')]

df = df.pivot(index = 'AREA_NAME', columns ='OCC_TITLE', values = 'TOT_EMP').fillna(0)

#sort occs by size 
sorted_occix = np.argsort(df.sum(0).values)

#sort MSA's by size 
sorted_MSAix = np.argsort(df.sum(1).values)

df = df.iloc[sorted_MSAix,sorted_occix]

major_df = df/10 #note that we divide by 10 the data, assuming it is generated per 10 
```

```python
#detailed occupation groups 
df = pd.read_csv('./BLS2016_clean.csv')

df = df.loc[df['OCC_GROUP']=='detailed']

#drop Farming and drop PR 
df = df[~df['OCC_CODE'].str.startswith('45')] # != 'Farming, Fishing, and Forestry'] 
df = df[~df['AREA_NAME'].str.contains('PR')]

df = df.pivot(index = 'AREA_NAME', columns ='OCC_TITLE', values = 'TOT_EMP').fillna(0)

#sort occs by size 
sorted_occix = np.argsort(df.sum(0).values)[::-1]

#sort MSA's by size 
sorted_MSAix = np.argsort(df.sum(1).values)[::-1]

df = df.iloc[sorted_MSAix,sorted_occix]

detailed_df = df.loc[:,df.sum(0)>df.sum(0).quantile(.1)]/10 #remove smallest ten percent occupations, and assume data is generated per 10 
```

## function for plotting PMI matrices for different priors 

```python
def make_pairplot_priors_coloc(df, perc):

    nr_prior_obs = df.sum().sum()*perc / (1-perc) 

    #for coloc

    Q_u = CoLoc(df, prior = 'uniform', nr_prior_obs = nr_prior_obs) 
    #Q_m = CoLoc(df, prior = 'maxent', nr_prior_obs = nr_prior_obs) 
    Q_p = CoLoc(df, prior = 'prop', nr_prior_obs = nr_prior_obs) 
    Q_n = CoLoc(df, prior = None)

    PMI_u = Q_u.make_PMIpij()
    #PMI_m = Q_m.make_PMIpij()
    PMI_p = Q_p.make_PMIpij()
    PMI_n = Q_n.make_PMIpij() 

    occ_zero = (df==0).sum(0)>0

    df = PMI_u.where(np.triu(np.ones(PMI_u.shape),1)==1).stack().to_frame('uniform').join(PMI_p.stack().to_frame('proportional')).join(PMI_n.stack().to_frame('zero prior'))#.join(PMI_m.stack().to_frame('maxent'))
    df = df.join((Q_n.pij==0).stack().to_frame('zero')).sort_values('zero', ascending = True)
    df['zero'] = df['zero'].apply(lambda x: 'zero value' if x else 'nonzero value')
    
    df.index = df.index.set_names(['occupation 1', 'occupation 2'])

    g = sns.PairGrid(df, corner = True, hue = 'zero', hue_order = ['nonzero value', 'zero value'])

    g.map_lower(sns.scatterplot, alpha = .3)

    g.map_diag(sns.histplot, log_scale = [None,10], alpha = .5)
    g.add_legend(bbox_to_anchor=(.65, .75), loc='upper left', title = '')

    g.fig.suptitle(f'Co-location associations, prior weight {perc:.1E}')
```

```python

def make_pairplot_priors_loc(df, perc):
    ## perc is prior weight(percentage) 
    ##df is count data 
    
    nr_prior_obs = df.sum().sum()*perc / (1-perc) 
    
    Q_u = CoLoc(df, prior = 'uniform', nr_prior_obs = nr_prior_obs) 
    #Q_m = CoLoc(df, prior = 'maxent', nr_prior_obs = nr_prior_obs) 
    Q_p = CoLoc(df, prior = 'prop', nr_prior_obs = nr_prior_obs) 
    Q_n = CoLoc(df[df>0], prior = 'prop', nr_prior_obs = nr_prior_obs) 

    PMI_u = Q_u.make_PMIpci()
    #PMI_m = Q_m.make_PMIpci()
    PMI_p = Q_p.make_PMIpci()
    PMI_n = Q_n.make_PMIpci()

    zeros = (df==0)

    df = PMI_u.stack().to_frame('uniform').join(PMI_p.stack().to_frame('proportional')).join(PMI_n.stack().to_frame('zero prior'))#.join(PMI_m.stack().to_frame('maxent'))
    df = df.join(zeros.stack().to_frame('zero')).sort_values('zero', ascending = False)
    df['zero'] = df['zero'].apply(lambda x: 'zero value' if x else 'nonzero value')
    
    df.index = df.index.set_names(['occupation 1', 'occupation 2'])

    g = sns.PairGrid(df, corner = True, hue = 'zero', hue_order = ['nonzero value', 'zero value'])

    g.map_lower(sns.scatterplot, alpha = .3)

    g.map_diag(sns.histplot, log_scale= [None,10], alpha = .5)
    g.add_legend(bbox_to_anchor=(.65, .75), loc='upper left', title = '')

    
    g.fig.suptitle(f'Location associations, prior weight {perc:.1E}')
```

```python
#for colocation associations 
for perc in [.0007,.05,.5]:
    plt.figure()
    make_pairplot_priors_coloc(major_df, perc)
    plt.savefig(f'./figures/prior_pairplot_coloc_{perc}.png', bbox_inches = 'tight', dpi=300)
```

```python
# for location associations 
for perc in [.0007,.05,.5]:
    plt.figure()
    make_pairplot_priors_loc(major_df, perc)
    plt.savefig(f'./figures/prior_pairplot_loc_{perc}.png', bbox_inches = 'tight', dpi=300)
```
